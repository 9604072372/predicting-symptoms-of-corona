```python
xtrain,xtest,ytrain,ytest = train_test_split(x_sm,y_sm,test_size=0.3,random_state=0)
```

```python
lr=LogisticRegression()
svc=SVC()
dt=DecisionTreeClassifier()
nb = MultinomialNB()
abc=AdaBoostClassifier()
bc=BaggingClassifier()
kc=KNeighborsClassifier()
rfc=RandomForestClassifier()
```

```python
# KNeighborsClassifier
lr.fit(xtrain,ytrain)
lr_ypred=lr.predict(xtest)
lr_data = pd.DataFrame({'Actual':ytest,'predicted':lr_ypred})
print(lr_data)
print(lr.score(xtrain,ytrain))
lr_acc=accuracy_score(ytest,lr_ypred)
print(lr_acc)
print(classification_report(ytest,lr_ypred))
cm=confusion_matrix(ytest,lr_ypred)
```
